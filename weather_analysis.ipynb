# Weather Analysis

## Project Overview

This project aims to analyze the daily climate data of Delhi from 2013 to 2017, focusing on changes in temperature, humidity, wind speed, and mean pressure. The dataset used is the Daily Climate Data in the City of Delhi from Kaggle.


## Set Up

**Data Collection and Preprocessing**

Dataset: Daily Climate Data in the City of Delhi from 2013 to 2017 - Kaggle


```python
# importing necessary library

import pandas as pd

# load dataset

df = pd.read_csv('./DailyDelhiClimateTrain.csv')

# check dataset

print(df.head())
print(df.info())
```

## Data Visualization

**Data Visualization**

- Daily Mean Temperature
- Daily Humidity
- Daily Wind Speed
- Daily Mean Pressure

```python
# importing the necessary library

import matplotlib.pyplot as plt

# visualize daily average temperature 

plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['meantemp'], label = 'Mean Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Daily Mean Temperature in Delhi (2013-2017)')
plt.legend()
plt.show()


# visualize humidity

plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['humidity'], label= 'Humidity', color = 'b')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.title('Daily Humidity in Delhi (2013-2017)')
plt.legend()
plt.show()


# visualize wind_speed

plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['wind_speed'], label= 'Wind Speed', color = 'g')
plt.xlabel('Date')
plt.ylabel('Wind Speed (km/h)')
plt.title('Daily Wind Speed in Delhi (2013-2017)')
plt.legend()
plt.show()


# Visualize mean pressure
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['meanpressure'], label='Mean Pressure', color='r')
plt.xlabel('Date')
plt.ylabel('Mean Pressure (hPa)')
plt.title('Daily Mean Pressure in Delhi (2013-2017)')
plt.legend()
plt.show()
```

## Statistical Analysis

### Descriptive Statistics

- Basic Statistics
    - Analyzed mean, variance, maximum, and minimum statistics for temperature.
    - Conducted similar statistical analysis for humidity, wind speed, and mean pressure.

```python
# stats for temperature

print(df['meantemp'].describe())

# stats for humidity
print(df['humidity'].describe())

# stats for wind speed
print(df['wind_speed'].describe())

# stats for mean air pressure
print(df['meanpressure'].describe())
```

### Corelation Analysis

- Analyzed the correlation between temperature, humidity, wind speed, and mean pressure.

```python
correlation = df[['meantemp', 'humidity', 'wind_speed', 'meanpressure']].corr()
print(correlation)
```

### Time Series Analysis
- Performed time series decomposition on temperature data to analyze trend, seasonality, and residuals.

```python
# importing necessary library

! pip install statsmodels

from statsmodels.tsa.seasonal import seasonal_decompose

# time series decompostion

result = seasonal_decompose(df['meantemp'], model='multiplicative', period = 365)
result.plot()
plt.show()
```
